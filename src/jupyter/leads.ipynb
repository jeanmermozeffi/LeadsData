```python
def filter_clients(dataframe: DataFrame, clients_list: list = default_clients_list, threshold: int = 95) -> DataFrame:
    def is_client(company_name: str) -> bool:
        for client in clients_list:
            if fuzz.partial_ratio(company_name.lower(), client.lower()) >= threshold:
                return True
        return False

    return dataframe[dataframe['current_company_name'].apply(is_client)]
```

```python
filtered_no_client_df = filter_clients(df)
```
